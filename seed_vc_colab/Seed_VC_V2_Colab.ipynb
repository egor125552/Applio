# Seed-VC V2: голос в голос без обучения

1. Включи **T4 GPU**: `Среда выполнения → Сменить среду выполнения → T4 GPU`.
2. Нажми **Выполнить всё**.
3. В последней ячейке дождись публичной ссылки `gradio.live` и открой её.
4. В `Source Audio` загрузи исходную речь, в `Reference Audio` — чистый пример нужного голоса длиной 10–25 секунд.

Обучение модели не требуется. Для первого запуска оставь параметры интерфейса по умолчанию.

```python
from pathlib import Path
import subprocess
import sys
import urllib.request

TOOLS = Path('/content/seed-vc-colab-tools')
TOOLS.mkdir(parents=True, exist_ok=True)
BASE = 'https://raw.githubusercontent.com/egor125552/Applio/agent/seed-vc-colab/seed_vc_colab'

for filename in ('install_seed_vc.py', 'requirements-colab.txt', 'launch_seed_vc.py'):
    urllib.request.urlretrieve(f'{BASE}/{filename}', TOOLS / filename)

subprocess.run([sys.executable, str(TOOLS / 'install_seed_vc.py')], check=True)
```
